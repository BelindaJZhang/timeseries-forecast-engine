```python
import os
import pandas as pd

df = pd.read_excel(r"C:\Users\juang\Desktop\bootcamp\capstone_project\bkw-energie\data\LNG Project data\DATA from 2009 (on progress)/Price_inv_stor_LNG_water_pipe_TTFMc_raw_FIXED.xlsx")

# Ensure Date column is datetime
df["Date"] = pd.to_datetime(df["Date"])

# Set index for interpolation
df = df.set_index("Date")

# Interpolate time-series (linear across dates)
df_interp = df.interpolate(method="time")

# Extract each series
ttf = df_interp[["TTF_Price"]].dropna()
pvb = df_interp[["PVB_Price"]].dropna()
jkm = df_interp[["JKM_Price"]].dropna()

# Ensure output folder exists
outdir = "data"
os.makedirs(outdir, exist_ok=True)

# Save files
ttf.to_csv(f"{outdir}/TTF_price.csv")
pvb.to_csv(f"{outdir}/PVB_price.csv")
jkm.to_csv(f"{outdir}/JKM_price.csv")
```
